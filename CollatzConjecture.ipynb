```python
c_100 = [*range(10,101,1)]
for number in c_100:
  a = number
  collatz = [a]
  while a != 1:
    if a%2 == 0:
      a= a//2
      collatz.append(a)
    else:
      a= a * 3 +1
      collatz.append(a)
  print(f'{collatz[0]}, {collatz[-3:]}')
```
